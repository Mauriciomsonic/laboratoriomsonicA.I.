```python
# Instalar e importar bibliotecas
!pip install gtts
from gtts import gTTS
from IPython.display import Audio, display
import os

# Texto para converter em áudio
texto = """
O curso 'dio.me bairesdev-machine-learning-training' é uma iniciativa excepcional para quem deseja mergulhar no universo do Machine Learning e LLMs. A estrutura do bootcamp é completa e prática, combinando fundamentos teóricos com projetos hands-on que realmente preparam o profissional para o mercado.

Os pontos mais destacados incluem:

Mentorias com experts: O contato direto com profissionais da BairesDev proporciona insights valiosos sobre carreira e tecnologia.

Desafios de código e projeto: Os exercícios práticos, como o sistema de assistência virtual com PLN, permitem aplicar imediatamente os conceitos aprendidos.

Abordagem em PLN: O conteúdo sobre Processamento de Linguagem Natural é especialmente relevante, cobrando desde conceitos básicos até implementações avançadas com reconhecimento de voz e síntese de fala.

Ferramentas de aprendizado: O uso de fóruns, rooms e rankings incentiva a colaboração e o networking entre os participantes.

Recomendo fortemente este curso para quem busca uma formação robusta e atualizada em Machine Learning, com foco em aplicações reais e valorização do portfólio profissional.
"""

# Converter texto em áudio usando gTTS (Google Text-to-Speech)
tts = gTTS(text=texto, lang='pt', slow=False)
tts.save("avaliacao_curso.mp3")

# Reproduzir o áudio diretamente no Colab
display(Audio("avaliacao_curso.mp3", autoplay=True))

# Opcional: Baixar o arquivo
from google.colab import files
files.download("avaliacao_curso.mp3")

# Opcional: Remover o arquivo após reproduzir (comentado para evitar erro durante o download)
# os.remove("avaliacao_curso.mp3")

```
